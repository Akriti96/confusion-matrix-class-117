```python
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix,classification_report
import pandas as pd
import seaborn as sns
```

```python
actual_data = ["Dog","Not a dog","Dog","Dog",      "Dog", "Not a dog", "Not a dog", "Dog",       "Dog", "Not a dog"]
predicted_data = ["Dog","Dog",      "Dog","Not a dog","Dog", "Not a dog", "Dog",       "Not a dog", "Dog", "Dog"]
labels = ["Dog", "Not a dog"]
cm = confusion_matrix(actual_data, predicted_data, labels)
```

```python
ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax)
ax.set_xlabel('Predicted')
ax.set_ylabel('Actual') 
# ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(labels); ax.yaxis.set_ticklabels(labels)
```
